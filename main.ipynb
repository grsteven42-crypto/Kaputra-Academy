```python
import sqlite3

db_path = r"D:\Documents\GitHub\Kaputra-Academy\prisma\dev.db"

conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Hapus tabel lama
cursor.execute("DROP TABLE IF EXISTS MockSubmission")

# Buat tabel baru
cursor.execute("""
CREATE TABLE MockSubmission (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    questionType TEXT NOT NULL,
    question TEXT NOT NULL,
    image TEXT,
    answer TEXT NOT NULL,
    createdAt DATETIME DEFAULT CURRENT_TIMESTAMP
)
""")

conn.commit()
conn.close()

print("Table berhasil diupdate.")
```
